```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
import os
import sys
sys.path.insert(0, os.path.abspath('./'))
import gridworld as gw
```

# List all available models

```python
model_filenames = os.listdir("models/")
model_filenames
```

# Pick a model and load it

```python
model = model_filenames[0]
data = gw.load_data(model)
```

# Pick a mission from the data set

```python
id = 20
```

# Compute the boundary strength profiles

```python
# INFORMATION CONTENT
ic = gw.ic(data,id)

# INFORMATION GAIN
ig = gw.ig(data,id)

# ENTROPY
en = gw.en(data,id)

# ACTION SEQUENCE
acns = gw.action(data,id)

# PICKUP POINT
p = gw.pickup(data,id)

# CHANGE IN INFORMATION CONTENT
ic_diff = gw.ic_diff(data,id)

# CHANGE IN ENTROPY
en_diff = gw.en_diff(data,id)

# CHANGE IN INFORMATION GAIN
ig_digg = gw.ig_diff(data,id)

# ENTROPY / INFORMATION CONTENT
en_ic = gw.en_ic(data,id)
```

# Estimate boundary location from profiles

```python
# MINIMUM INFORMATION CONTENT
min_ic = gw.min_ic(data,id)

# MINIMUM ENTROPY
min_en = gw.min_en(data,id)

# MAXIMUM CHANGE IN ENTROPY
max_en_diff = gw.max_en_diff(data,id)

# MAXIMUM CHANGE IN INFORMATION CONTENT 
max_id_diff = gw.max_ic_diff(data,id)

# MAXIMUM ENTROPY / INFORMATION CONTENT
max_en_ic = gw.max_en_ic(data,id)

# MAXIMUM CHANGE IN INFORMATION GAIN
max_ig_diff = gw.max_ig_diff(data,id)
```

# Visualise the results

```python
plt.figure(figsize=(15,5))
gw.plot(ic,'ic')
gw.plot(en,'en')
gw.plot(ig,'ig')
#gw.plot(ic_diff,'ic_diff')
#gw.plot(en_diff,'en_diff')
gw.mark(p,c='r',linestyle='--',label='pickup')
#gw.mark(min_ic,label='min_ic')
#gw.mark(max_en_diff,label='max en diff')
#gw.mark(max_ic_diff,label='max ic diff')
gw.mark(max_ig_diff,label='max ig diff')
#gw.mark(max_en_ic,label='max h/ic')
gw.set_xaxis(acns)
```
