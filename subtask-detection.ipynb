# Sub-task boundary detection

```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
import os
import sys
sys.path.insert(0, os.path.abspath('./'))
import gridworld as gw
```

```python
model_names = gw.list_model_names()

estimator_names = [
    "min_ic",
    "min_en",
    "max_en_diff",
    "max_en_diff_minus1",
    "max_ic_diff",
    "max_ic_diff_minus1",
    "max_ig_diff",
    "max_ig_diff_minus1",
    "max_en_ic",
    "max_ig_ic"
]

models = [gw.load_data(model_name) for model_name in model_names]
estimators = [getattr(gw,estimator_name) for estimator_name in estimator_names]
```

The following table contains numbers of correct subtask boundaries detected (out of 1000) for each model (columns) and estimator (rows). 

```python
results = {}

for m,M in enumerate(model_names):
    results[M] = [gw.eval_model(models[m],e) for e in estimators]
```

```python
pandas.DataFrame(data=results,index=estimator_names).transpose()
```
