```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
import os
import sys
sys.path.insert(0, os.path.abspath('./'))
import gridworld as gw
```

# List all models and estimators

```python
# GET THE NAMES OF AVAILABLE MODELS
model_names = gw.list_model_names()

# LIST ESTIMATORS 
## (UN)COMMENT TO SELECT REQUIRED
estimator_names = [
    "min_ic",
    #"min_en",
    #"min_ig",
    #"max_ic",
    "max_en",
    "max_ig",
    #"max_ic_diff",
    "max_en_diff",
    "max_ig_diff",
    #"min_ic_diff",
    #"min_en_diff",
    #"min_ig_diff",
    "max_ic_diff_minus1",
    #"max_en_diff_minus1",
    #"max_ig_diff_minus1",
    "max_en_ic",
    "max_ig_ic"
]

models = [gw.load_data(model_name) for model_name in model_names]
estimators = [getattr(gw,estimator_name) for estimator_name in estimator_names]
```

The following table contains numbers of correct subtask boundaries detected (out of 1000) for each model (rows) and estimator (columns). 

```python
results = {}

for m,M in enumerate(model_names):
    results[M] = [gw.eval_model(models[m],e) for e in estimators]
```

```python
pandas.DataFrame(data=results,index=estimator_names).transpose()
```
