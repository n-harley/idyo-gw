# Model: idyom-select-maxlinks3-minus

Idyom was run on the gridworld data using the following command:

`(idyom:idyom 33 '(action) '((action canseedoor)(action canseedoor canseekey)(action agentleft canseekey)(action agentleft canseedoor)(action agentright canseekey)(action agentright canseedoor)(action agentinfront canseekey)(action agentinfront canseedoor)(action agentinfront agentleft)(action agentinfront agentright)(action agentquad keyquad)))`

This uses the following viewpoints:

- `(ACTION CANSEEDOOR)`
- `(ACTION CANSEEDOOR CANSEEKEY)`
- `(ACTION AGENTLEFT CANSEEKEY)`
- `(ACTION AGENTLEFT CANSEEDOOR)`
- `(ACTION AGENTRIGHT CANSEEKEY)`
- `(ACTION AGENTRIGHT CANSEEDOOR)`
- `(ACTION AGENTINFRONT CANSEEKEY)`
- `(ACTION AGENTINFRONT CANSEEDOOR)`
- `(ACTION AGENTINFRONT AGENTLEFT)`
- `(ACTION AGENTINFRONT AGENTRIGHT)`
- `(ACTION AGENTQUAD KEYQUAD)`

These are the viewpoints selected in [idyom-select-maxlinks3](./idyom-select-maxlinks3.ipynb) minus the ones whose symmetric counterparts were not included.

The average IC was 0.68368334

```python
filename = 'idyom-select-maxlinks3-minus.csv'
data = gw.load_data(filename)
```

```python
id = 120
```

```python
# INFORMATION CONTENT,ic
ic = gw.ic(data,id)

# INFORMATION GAIN, ig
ig = gw.ig(data,id)

# ENTROPY, h
h = gw.h(data,id)

# ACTION SEQUENCE, acns
acns = gw.action(data,id)

# PICKUP POINT, p
p = gw.pickup(data,id)

# CHANGE IN INFORMATION CONTENT, dic
dic = gw.ic_diff(data,id)

# CHANGE IN ENTROPY, dh
dh = gw.h_diff(data,id)

# ENTROPY / INFORMATION CONTENT
hic = gw.h_ic(data,id)

# MINIMUM INFORMATION CONTENT
minic = gw.min_ic(data,id)

# MINIMUM ENTROPY
minh = gw.min_h(data,id)

# MAXIMUM CHANGE IN ENTROPY, max_dh
mdh = gw.max_h_diff(data,id)

# MAXIMUM CHANGE IN INFORMATION CONTENT, max_dic
mdic = gw.max_ic_diff(data,id)

# MAXIMUM ENTROPY / INFORMATION CONTENT, mhic
mhic = gw.max_h_ic(data,id)
```

```python
plt.figure(figsize=(15,5))
gw.plot(ic,'ic')
gw.plot(h,'h')
#gw.plot(ig,'ig')
#gw.plot(dic,'dic')
#gw.plot(dh,'dh')
gw.mark(p,c='r',linestyle='--',label='pickup')
gw.mark(minic,label='min ic')
gw.mark(mdh,label='max h diff')
gw.mark(mdic,label='max ic diff')
gw.mark(mhic,label='max h/ic')
gw.set_xaxis(acns)
```
