```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
import os
import sys
sys.path.insert(0, os.path.abspath('./'))
import gridworld as gw
```

Idyom was run on the gridwold data using the following command:

(idyom:idyom 33 '(action) :select :basis '(action orientation agentinfront agentbehind agentleft agentright agentx agenty agentquad keyquad doorquad agentinfrontleft agentinfrontright agentbehindleft agentbehindright canseedoor canseekey distance) :max-links 2)

The program considers viewpoints from the basis set as well as binary and ternary linked combination. It searches for the combination of these viewpoints which gives the most compact model. That is, is tries to minimise the mean information content by a simple hill-climbing algorithm.

The set of viewpoints selected by the program was as follows:

- (ACTION AGENTINFRONT)
- (ACTION AGENTRIGHT) 
- (ACTION AGENTLEFT) 
- (ACTION AGENTINFRONTLEFT)
- (ACTION AGENTINFRONTRIGHT) 
- (ACTION CANSEEDOOR)
- (ACTION DISTANCE)

The mean information content of the resulting model is 0.7807406.

```python
filename = 'idyom-select-maxlinks2.csv'
data = gw.load_data(filename)
```

```python
id = 20
```

```python
# INFORMATION CONTENT,ic
ic = gw.ic(data,id)

# INFORMATION GAIN, ig
ig = gw.ig(data,id)

# ENTROPY, h
h = gw.h(data,id)

# ACTION SEQUENCE, acns
acns = gw.action(data,id)

# PICKUP POINT, p
p = gw.pickup(data,id)

# CHANGE IN INFORMATION CONTENT, dic
dic = gw.ic_diff(data,id)

# CHANGE IN ENTROPY, dh
dh = gw.h_diff(data,id)

# ENTROPY / INFORMATION CONTENT
hic = gw.h_ic(data,id)

# MINIMUM INFORMATION CONTENT
minic = gw.min_ic(data,id)

# MINIMUM ENTROPY
minh = gw.min_h(data,id)

# MAXIMUM CHANGE IN ENTROPY, max_dh
mdh = gw.max_h_diff(data,id)

# MAXIMUM CHANGE IN INFORMATION CONTENT, max_dic
mdic = gw.max_ic_diff(data,id)

# MAXIMUM ENTROPY / INFORMATION CONTENT, mhic
mhic = gw.max_h_ic(data,id)
```

```python
plt.figure(figsize=(15,5))
gw.plot(ic,'ic')
gw.plot(h,'h')
#gw.plot(ig,'ig')
#gw.plot(dic,'dic')
#gw.plot(dh,'dh')
gw.mark(p,c='r',linestyle='--',label='pickup')
gw.mark(minic,label='min ic')
gw.mark(mdh,label='max h diff')
gw.mark(mdic,label='max ic diff')
gw.mark(mhic,label='max h/ic')
gw.set_xaxis(acns)
```
