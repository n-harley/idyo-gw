# Model: idyom-select-maxlinks2-plus

Idyom was run on the gridwold data using the following command:

`(idyom:idyom 33 '(action) '((action agentinfront)(action agentright)(action agentleft)(action agentinfrontleft)(action agentinfrontleft)(action canseedoor)(action canseekey)(action distance)))`

The set of viewpoints used were as follows:

- `(ACTION AGENTINFRONT)`
- `(ACTION AGENTRIGHT)`
- `(ACTION AGENTLEFT)`
- `(ACTION AGENTINFRONTLEFT)`
- `(ACTION AGENTINFRONTRIGHT)` 
- `(ACTION CANSEEDOOR)`
- `(ACTION CANSEEKEY)`
- `(ACTION DISTANCE)`

This uses the viewpoints selected in [idyom-select-maxlinks2](./idyom-select-maxlinks2.ipynb) plus the only ommited first-person viewpoint `CANSEEKEY`.  

The mean information content of the resulting model is 0.7917453.

```python
filename = 'idyom-select-maxlinks2-plus.csv'
data = gw.load_data(filename)
```

```python
id = 20
```

```python
# INFORMATION CONTENT,ic
ic = gw.ic(data,id)

# INFORMATION GAIN, ig
ig = gw.ig(data,id)

# ENTROPY, h
h = gw.h(data,id)

# ACTION SEQUENCE, acns
acns = gw.action(data,id)

# PICKUP POINT, p
p = gw.pickup(data,id)

# CHANGE IN INFORMATION CONTENT, dic
dic = gw.ic_diff(data,id)

# CHANGE IN ENTROPY, dh
dh = gw.h_diff(data,id)

# ENTROPY / INFORMATION CONTENT
hic = gw.h_ic(data,id)

# MINIMUM INFORMATION CONTENT
minic = gw.min_ic(data,id)

# MINIMUM ENTROPY
minh = gw.min_h(data,id)

# MAXIMUM CHANGE IN ENTROPY, max_dh
mdh = gw.max_h_diff(data,id)

# MAXIMUM CHANGE IN INFORMATION CONTENT, max_dic
mdic = gw.max_ic_diff(data,id)

# MAXIMUM ENTROPY / INFORMATION CONTENT, mhic
mhic = gw.max_h_ic(data,id)
```

```python
plt.figure(figsize=(15,5))
gw.plot(ic,'ic')
gw.plot(h,'h')
#gw.plot(ig,'ig')
#gw.plot(dic,'dic')
#gw.plot(dh,'dh')
gw.mark(p,c='r',linestyle='--',label='pickup')
gw.mark(minic,label='min ic')
gw.mark(mdh,label='max h diff')
gw.mark(mdic,label='max ic diff')
gw.mark(mhic,label='max h/ic')
gw.set_xaxis(acns)
```
